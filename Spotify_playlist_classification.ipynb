```python
import tensorflow as tf
import numpy as np 
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler


```

```python
spotify2 = pd.read_csv('fooh.csv')
spotify2.pop('type')
spotify2.pop('id')
spotify2.pop('uri')
spotify2.pop('track_href')
spotify2.pop('analysis_url')
spotify2.head()
```

```python
spotify1 = pd.read_csv('foo.csv')
spotify1.pop('type')
spotify1.pop('id')
spotify1.pop('uri')
spotify1.pop('track_href')
spotify1.pop('analysis_url')
spotify1.head()
```

```python
spotify2['R&B'] = np.zeros ((17,))
spotify1['R&B'] = np.ones(shape = 17)
spotify = pd.merge(spotify1,spotify2, how='outer')
spotify.info()
```

```python
scaler = MinMaxScaler()
rspotify = scaler.fit_transform(spotify)
spotify = pd.DataFrame(rspotify, columns =['unnamed: 0','danceability','energy', 'key','loudness ','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo','duration_ms','time_signature','R&B'])
y = spotify.pop('R&B')
```

```python
p = []
for i in range(0,len(y)):
    if y[i] == 0:
        p.append([1,0])
    else:
        p.append([0,1])
print(p)

df1 = pd.DataFrame (p, columns = ['Playlist f00', 'Playlist f00h'])

tensor_y = tf.convert_to_tensor(y)
tensor_spotify = tf.convert_to_tensor(spotify)
 
    
```

```python
x_train,x_test,y_train,y_test=train_test_split(spotify,df1,test_size=0.3)
```

```python
print(x_train)
```

```python
model = Sequential()
model.add(Dense(150,activation='relu',input_dim = 14))
model.add(Dense(150,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(150,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(2,activation='softmax'))



model.compile(optimizer='Adam',loss= 'categorical_crossentropy',metrics = ['accuracy'])
model.fit(x_train,y_train,epochs= 200,batch_size= 200)




```

```python
y_pred = model.predict(x_test)
score = model.evaluate(x_test, y_test,verbose=1)

print(score)
```
